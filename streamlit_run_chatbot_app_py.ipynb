```python
import streamlit as st
from transformers import pipeline

# Configurazione chatbot
@st.cache_resource
def load_chatbot():
    return pipeline("text-generation", model="gpt2")

chatbot = load_chatbot()

# Configurazione della pagina
st.set_page_config(page_title="Chatbot con Temperatura", layout="wide")

# Sidebar per impostare la temperatura
st.sidebar.header("Impostazioni Chatbot")
temperature = st.sidebar.slider("Temperatura (creatività):", min_value=0.0, max_value=1.5, value=0.7, step=0.1)

# Titolo nella parte centrale
st.title("Chatbot Creativo con Temperatura Regolabile")
st.write("Chatta con il chatbot e modifica la sua creatività regolando la temperatura dalla barra laterale.")

# Chat
if "messages" not in st.session_state:
    st.session_state.messages = []

# Form di input
with st.form("chat_form"):
    user_input = st.text_input("Tu:", key="user_input")
    submit_button = st.form_submit_button("Invia")

# Logica della chat
if submit_button and user_input:
    # Aggiungi il messaggio dell'utente
    st.session_state.messages.append({"role": "user", "content": user_input})

    # Risposta del chatbot
    chatbot_response = chatbot(user_input, max_length=50, do_sample=True, temperature=temperature)[0]['generated_text']
    st.session_state.messages.append({"role": "bot", "content": chatbot_response})

# Mostra la conversazione
for msg in st.session_state.messages:
    if msg["role"] == "user":
        st.write(f"**Tu:** {msg['content']}")
    else:
        st.write(f"**Chatbot:** {msg['content']}")
```
